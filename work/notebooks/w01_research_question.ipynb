## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**Question:** Which visible, well-ranked pages are under-capturing clicks relative to other pages at the same search position, and would benefit from a title/meta/content review?

**Decision improved:** Which pages a content editor should prioritize reviewing for title/meta/snippet improvements.

**Who acts, and how:** A FlyRank content reviewer opens the ranked list and manually reviews/rewrites the top candidates' titles and meta descriptions.

**Cost of a wrong call:** A false positive wastes limited editor time reviewing a page that was actually performing normally for its position. A false negative means a page keeps under-converting clicks it could otherwise be getting, silently costing traffic. Neither is catastrophic — this is a prioritization tool, not an automated action.
Why ML/data, not a simple rule: A flat CTR threshold ignores position — 2% CTR is excellent at position 20 but poor at position 3. This requires comparing each page against its position-tier peers, which needs grouped statistics rather than one universal cutoff.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

import os, sys, subprocess

IN_COLAB = "google.colab" in sys.modules
REPO_URL = "https://github.com/widadfatimakhan/flyrank-internship-ml"
REPO_DIR = "flyrank-internship-ml"

if IN_COLAB:
    if not os.path.isdir(REPO_DIR):
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)
    os.chdir(REPO_DIR)

print("Working dir:", os.getcwd())
assert os.path.exists("data/raw/content_refresh_anonymized.csv"), "starter CSV not found — check the clone worked"

import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# avg_position == 0 means "no position data," not rank zero -- exclude it
has_position = df["avg_position"] > 0
visible = has_position & (df["impressions_90d"] >= 500)
vis_df = df[visible].copy()

# Expected CTR per position tier, then each page's gap vs its own tier's average
tier_avg_ctr = vis_df.groupby("position_tier")["ctr"].transform("mean")
vis_df["ctr_gap"] = tier_avg_ctr - vis_df["ctr"]  # positive = underperforming its tier

underperformers = vis_df[vis_df["ctr_gap"] > 0]

print(f"Visible pages with real position data (impressions_90d >= 500): {len(vis_df):,}")
print(f"Pages under-capturing clicks vs. their own position tier's average: {len(underperformers):,}")
print(f"Share of visible pages that are under-capturing: {len(underperformers)/len(vis_df):.1%}")
print(f"Average CTR gap among underperformers: {underperformers['ctr_gap'].mean():.3f} (in ctr's own units)")

# Observations: Among 16,726 pages with real position data and meaningful traffic (≥500 impressions/90 days),
# 65.3% (10,920 pages) have a CTR below their own position tier's average — with an average gap of 0.152
# percentage points among underperformers.
# This is a meaningful pool of pages to prioritize for review.
# Note: because roughly half of any group naturally falls below its own average,
# this high share suggests the CTR distribution within each tier is right-skewed
# (a smaller number of high performers pulling the average up),
# which I'd want to look into further rather than over-read the 65% figure at face value.
```
